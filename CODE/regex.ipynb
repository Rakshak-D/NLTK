```python
import re
```

```python
text="The rain in Spain"
test1=re.search("^The.*Spain$",text)
```

```python
print(test1.start())
```

```python
if test1:
    print("Match has been found")
else:
    print("No match found .")
```

```python
test2=re.search("^The.*India$",text)
```

```python
if test2:
    print("Match has been found")
else:
    print("No match found .")
```

| Set         | Description                                                                 |
|-------------|-----------------------------------------------------------------------------|
| `[arn]`     | Returns a match where one of the specified characters (`a`, `r`, or `n`) is present |
| `[a-n]`     | Returns a match for any lowercase character alphabetically between `a` and `n` |
| `[^arn]`    | Returns a match for any character **except** `a`, `r`, and `n` |
| `[0123]`    | Returns a match where any of the specified digits (`0`, `1`, `2`, or `3`) are present |
| `[0-9]`     | Returns a match for any digit between `0` and `9` |
| `[0-5][0-9]`| Returns a match for any two-digit number from `00` to `59` |
| `[a-zA-Z]`  | Returns a match for any character alphabetically between `a` and `z`, lowercase or uppercase |
| `[+]`       | In sets, `+`, `*`, `.`, `|`, `()`, `$`, `{}` have no special meaning, so `[+]` matches a literal `+` character |
| Character | Description | Example |
|-----------|------------|----------|
| `\A` | Returns a match if the specified characters are at the beginning of the string | `"\AThe"` |
| `\b` | Returns a match where the specified characters are at the beginning or end of a word (the `r` prefix makes it a raw string) | `r"\bain"`<br>`r"ain\b"` |
| `\B` | Returns a match where the specified characters are present, but NOT at the beginning or end of a word (the `r` prefix makes it a raw string) | `r"\Bain"`<br>`r"ain\B"` |
| `\d` | Returns a match where the string contains digits (0–9) | `"\d"` |
| `\D` | Returns a match where the string does NOT contain digits | `"\D"` |
| `\s` | Returns a match where the string contains a whitespace character | `"\s"` |
| `\S` | Returns a match where the string does NOT contain a whitespace character | `"\S"` |
| `\w` | Returns a match where the string contains any word character (`a–z`, `A–Z`, `0–9`, `_`) | `"\w"` |
| `\W` | Returns a match where the string does NOT contain any word character | `"\W"` |
| `\Z` | Returns a match if the specified characters are at the end of the string | `"Spain\Z"` |
