```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
sentence="NLTK (Natural Language Toolkit) is a leading open-source Python library used for Natural Language Processing (NLP) and computational linguistics."
```

```python
tokens=word_tokenize(sentence)
```

```python
print(tokens)
```

```python
sentence_tokens=sent_tokenize(sentence)
```

```python
print(sentence_tokens)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
porterstemmer=PorterStemmer()
```

```python
print(porterstemmer.stem("changing"))
print(porterstemmer.stem("change"))
print(porterstemmer.stem("changed"))
print(porterstemmer.stem("change"))
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
print(lemmatizer.lemmatize("changing","v"))
print(lemmatizer.lemmatize("change","v"))
print(lemmatizer.lemmatize("changed","v"))
print(lemmatizer.lemmatize("changes","v"))
```
